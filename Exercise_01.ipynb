```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Part a)

```python
# Defining the integrand as a function
def integrand(x, a, n):
    return x**n / (x + a)
```

```python
# Setting values for the plots
a = 5.
n = np.array([1, 5, 10, 20, 30, 50])
x = np.linspace(0, 1, 500)
```

```python
# Plotting in a for loop for all different n
plt.figure(figsize=(26,16), dpi=80)
for i in range(len(n)):
    plt.subplot(3, 2, i+1)
    plt.plot(x, integrand(x, a, n[i]))
    plt.xlabel('x')
    plt.ylabel('integrand')
    plt.title('Integrand for different values of n')
    plt.grid(b=True, which='major', color='#666666', linestyle='--', alpha = 0.8)
    plt.grid(b=True, which='minor', color='#666666', linestyle='--', alpha = 0.3)
    plt.minorticks_on()
```

#### Part b)

```python
# Setting values to test the function
a = 5
n0 = 0
y0 = 0.1823
n1 = 3
```

```python
# defining the numeric solution of the integral as a recursive function
def recursive_integral(a, y0, n0, n1):
    # if n1 is larger than n0, do a forward iteration
    if(n1>n0):
        y = (1/n1 - a*recursive_integral(a, y0, n0, n1 - 1))
        print(y)
        return y
    # if n1 is smaller than n0, do a backwards iteration
    if(n1<n0):
        y = (1/n1 - recursive_integral(a, y0, n0, n1 + 1))/a
        print(y)
        return y
    # if n1 has reached n0, return the known solution y0
    if(n1==n0):
        print(y0)
        return y0
```

```python
# Testing the function for example values
recursive_integral(a, y0, n0, n1)
```
